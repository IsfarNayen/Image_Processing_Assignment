Resizing the image by pillow library

```python
from PIL import Image
```

```python
img = Image.open('/kaggle/working/understanding-your-dog-s-body-language.jpg')
img_resized = img.resize((200 , 200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

**Converting RGB images to Grayscale by OpenCV**

```python
import cv2
```

```python
img = cv2.imread(pic)
```

```python
 img.shape
```

```python
type(img)
```

```python
bw_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
bw_img.shape
```

**Grayscale image shown by cv2 function, Computer Vision**

```python
from google.colab.patches import cv2_imshow
cv2_imshow(bw_img) #showing image by cv2
```

```python
#saving the grayscale image
cv2.imwrite('B&W_cv2.jpg' , bw_img)
```

**Loading image from Pillow(My contribution)**

```python
img = Image.fromarray(bw_img)
```

```python
img.save('dog_BlackandWhite.jpg')
```
